```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
```

```python
re=pd.read_excel(r'改变超参数.xlsx')
beta = chr(946)
gamma = chr(947)
```

```python
figure,(ax1,ax2)=plt.subplots(2,1,figsize=(8,6), dpi=100,sharex=True)

beta000005 = re['beta'] == 0.00005
beta00005 = re['beta'] == 0.0005
beta0005 = re['beta'] == 0.005
beta005 = re['beta'] == 0.05
beta05 = re['beta'] == 0.5

gamma_list = [0.85,0.90,0.95,0.99]

ax1.plot(gamma_list, re['rewards'][beta000005] ,c='#2A4373',ls='-',marker='o',label=f'{beta}=0.00005')
ax1.plot(gamma_list, re['rewards'][beta00005] ,c='#3400F5',ls=':',marker='v',label=f'{beta}=0.0005')
ax1.plot(gamma_list, re['rewards'][beta0005] ,c='#01F051',ls='--',marker='+',label=f'{beta}=0.005')
ax1.plot(gamma_list, re['rewards'][beta005] ,c='#F50B8C',ls='-.',marker='s',label=f'{beta}=0.05')
ax1.plot(gamma_list, re['rewards'][beta05] ,c='#A86A09',ls=(0,(3,1,1,1,1,1)),marker='*',label=f'{beta}=0.5')

ax2.plot(gamma_list, re['steps'][beta000005] ,c='#2A4373',ls='-',marker='o',label=f'{beta}=0.00005')
ax2.plot(gamma_list, re['steps'][beta00005] ,c='#3400F5',ls=':',marker='v',label=f'{beta}=0.0005')
ax2.plot(gamma_list, re['steps'][beta0005] ,c='#01F051',ls='--',marker='+',label=f'{beta}=0.005')
ax2.plot(gamma_list, re['steps'][beta005] ,c='#F50B8C',ls='-.',marker='s',label=f'{beta}=0.05')
ax2.plot(gamma_list, re['steps'][beta05] ,c='#A86A09',ls=(0,(3,1,1,1,1,1)),marker='*',label=f'{beta}=0.5')

ax1.set_ylabel('Average rewards')
ax2.set_ylabel('Average steps')
ax2.set_xlabel(f'Value of {gamma}')

ax1.set_xlim((0.845, 1.0))
ax1.set_ylim((0.90, 1))
ax2.set_ylim((3, 13))

x_major_locator=MultipleLocator(0.05)
# #把x轴的刻度间隔设置为1，并存在变量里
# ax=plt.gca()
# #ax为两条坐标轴的实例
ax1.xaxis.set_major_locator(x_major_locator)

ax2.legend()
figure.subplots_adjust(hspace=0)
figure.savefig("gamma_change.png")
```

```python
figure,(ax1,ax2)=plt.subplots(2,1,figsize=(8,6), dpi=100,sharex=True)

gamma85 = re['gamma'] == 0.85
gamma90 = re['gamma'] == 0.90
gamma95 = re['gamma'] == 0.95
gamma99 = re['gamma'] == 0.99

beta_list =  [0.00005,0.0005,0.005,0.05,0.5]
beta_list_lables = ['$5 \cdot 10^{-5}$',
                    '$5 \cdot 10^{-4}$',
                    '$5 \cdot 10^{-3}$',
                    '$5 \cdot 10^{-2}$',
                    '$5 \cdot 10^{-1}$']

ax1.semilogx(beta_list, re['rewards'][gamma85] ,c='#2A4373',ls='-',marker='o',label=f'{gamma}=0.85')
ax1.semilogx(beta_list, re['rewards'][gamma90] ,c='#3400F5',ls=':',marker='v',label=f'{gamma}=0.90')
ax1.semilogx(beta_list, re['rewards'][gamma95] ,c='#01F051',ls='--',marker='+',label=f'{gamma}=0.95')
ax1.semilogx(beta_list, re['rewards'][gamma99] ,c='#F50B8C',ls='-.',marker='s',label=f'{gamma}=0.99')

ax2.semilogx(beta_list, re['steps'][gamma85] ,c='#2A4373',ls='-',marker='o',label=f'{gamma}=0.85')
ax2.semilogx(beta_list, re['steps'][gamma90] ,c='#3400F5',ls=':',marker='v',label=f'{gamma}=0.90')
ax2.semilogx(beta_list, re['steps'][gamma95] ,c='#01F051',ls='--',marker='+',label=f'{gamma}=0.95')
ax2.semilogx(beta_list, re['steps'][gamma99] ,c='#F50B8C',ls='-.',marker='s',label=f'{gamma}=0.99')

ax1.set_ylabel('Average rewards')
ax2.set_ylabel('Average steps')
ax2.set_xlabel(f'Value of {beta}')

# ax1.set_xlim((0.845, 1.0))
ax1.set_ylim((0.90, 1))
ax2.set_ylim((3, 13))

ax2.set_xticks(beta_list,beta_list_lables)

ax1.legend()
figure.subplots_adjust(hspace=0)
figure.savefig("beta_change.png")
```

# gamma 越大收敛越不稳定

```python
def EMA(arr,period=1500):    ## Function for computing exponential moving average
    df = pd.DataFrame(arr)
    return df.ewm(span=period,min_periods=period).mean()

re08=pd.read_excel(r'result_hyperparameters_0.85_0.005.xlsx')
re09=pd.read_excel(r'result_hyperparameters_0.9_0.005.xlsx')
re10=pd.read_excel(r'result_hyperparameters_0.95_0.005.xlsx')
re11=pd.read_excel(r'result_hyperparameters_0.99_0.005.xlsx')
beta = chr(946)
gamma = chr(947)
N_episodes = 100000

figure,(ax1,ax2)=plt.subplots(2,1,figsize=(8,6), dpi=100,sharex=True)


ax1.plot(range(N_episodes), EMA(re08['R_save']),c='#2A4373',alpha=0.7,label=f'{gamma}=0.85, {beta}=0.005')
ax1.plot(range(N_episodes), EMA(re09['R_save']),c='#3400F5',alpha=0.7,label=f'{gamma}=0.90, {beta}=0.005')
ax1.plot(range(N_episodes), EMA(re10['R_save']),c='#01F051',alpha=0.7,label=f'{gamma}=0.95, {beta}=0.005')
ax1.plot(range(N_episodes), EMA(re11['R_save']),c='#F50B8C',alpha=0.7,label=f'{gamma}=0.99, {beta}=0.005')

ax2.plot(range(N_episodes), EMA(re08['N_moves_save']),c='#2A4373',alpha=0.7,label=f'{gamma}=0.85, {beta}=0.005')
ax2.plot(range(N_episodes), EMA(re09['N_moves_save']),c='#3400F5',alpha=0.7,label=f'{gamma}=0.90, {beta}=0.005')
ax2.plot(range(N_episodes), EMA(re10['N_moves_save']),c='#01F051',alpha=0.7,label=f'{gamma}=0.95, {beta}=0.005')
ax2.plot(range(N_episodes), EMA(re11['N_moves_save']),c='#F50B8C',alpha=0.7,label=f'{gamma}=0.99, {beta}=0.005')

ax1.set_ylabel('The reward per game')
ax2.set_ylabel('The number of moves per game')
ax2.set_xlabel('Training time')

ax1.legend()

figure.subplots_adjust(hspace=0)
figure.savefig("large_gamma_unstable.png")
```
